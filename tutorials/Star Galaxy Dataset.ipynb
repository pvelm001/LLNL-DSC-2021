How do we get an image out of this data? Here we have 31,798 images of 5410 columns.
The First column is junk, so lets take it out

```python
galstar = galstar.iloc[:,1:]
```

Next, lets get the y_values out so we know which label each value has

```python
galstar_labels = galstar['y']
```

Now, lets focus on getting images out of this data, there are 8, 26x26 images per object, 4 different bands, (think of these like the RBG Pixels) and 4 Point Spread Function Images

```python
galstar_images = galstar.iloc[:,1:]
```

```python
plt.imshow(np.reshape(galstar_images.iloc[0].values[0:int(5408/8)],(26,26)))
plt.show()
```

Lets Look at a sample star in all 4 bands and all corresponding PSFS

```python
for i in np.arange(8):
    plt.imshow(np.reshape(galstar_images.iloc[0].values[int(5408/8)*i:int(5408/8)*(i+1)],(26,26)))
    plt.show()
```
